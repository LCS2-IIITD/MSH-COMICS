```
EMBEDDING_DIM=300
embedding_matrix = np.zeros((len(word_index) + 1, EMBEDDING_DIM))
```

```
for word,i in word_index.items():
  embedding_vector=embeddings_dict.get(word)
  if embedding_vector is not None:
    embedding_matrix[i]=embedding_vector
```

```
embedding_matrix.shape
```

```
from tensorflow.keras.layers import Input, Dense,Conv1D,MaxPooling1D,Flatten,Dropout
```

```
from tensorflow.keras.layers import Embedding
```

```
embedding_layer=Embedding(len(word_index)+1,EMBEDDING_DIM,weights=[embedding_matrix],input_length=MAX_SEQUENCE_LENGTH,trainable=False,name='embedding',mask_zero=True)
```

```
sequence_input = Input(shape=(MAX_SEQUENCE_LENGTH,), dtype='int32')
embedded_sequences = embedding_layer(sequence_input)
```

```
embedding=[]
for i in range(len(df_data)):
  embedding.append(emb_dict[df_data['name'][i]])

df_data["embedding"]=embedding
```

```
fe_inputs = Input(shape=(2048,))
fe_layer1 = Dropout(0.4)(fe_inputs)
fe_layer2 = Dense(256, activation='relu')(fe_layer1)
seq_inputs = Input(shape=(MAX_SEQUENCE_LENGTH,))
seq_layer1 = embedding_layer(seq_inputs)

seq_layer2 = Dropout(0.4)(seq_layer1)
seq_layer3 = LSTM(256)(seq_layer2)
decoder_layer1 = add([fe_layer2, seq_layer3])
decoder_layer2 = Dense(128, activation='relu')(decoder_layer1)
outputs = Dense(1, activation='sigmoid')(decoder_layer2)
model = Model(inputs=[fe_inputs, seq_inputs], outputs=outputs)
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=["accuracy"])
```
